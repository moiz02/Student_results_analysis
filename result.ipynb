```python
# Creating an array of all the students roll numbers including lateral enteries
normal_students_lords = np.arange(160920733001, 160920733180)
lateral_entry_lords = np.arange(160920733301, 160920733320)

normal_students_mj = np.arange(160420733001, 160420733122)
lateral_entry_mj = np.arange(160420733301, 160420733313)

normal_students_deccan = np.arange(160320733001, 160320733091)
lateral_entry_deccan = np.arange(160320733301, 160320733310)
```

```python
# Combine roll numbers for each group into separate arrays
total_students_lords = np.concatenate((normal_students_lords, lateral_entry_lords))
total_students_mj = np.concatenate((normal_students_mj, lateral_entry_mj))
total_students_deccan = np.concatenate((normal_students_deccan, lateral_entry_deccan))

# Combine roll numbers of all groups into a single array
total_students = np.concatenate((total_students_deccan, total_students_mj, total_students_lords))
```

```python
# Define discontinued roll numbers
discontinued_students = np.array([160920733156, 160920733176, 160320733078])

# Remove discontinued roll numbers
total_students = np.setdiff1d(total_students, discontinued_students)
```
