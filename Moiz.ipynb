```python
# Read from the CSV file with the 'Roll Number' column as the index
df = pd.read_csv('results.csv', index_col='Roll Number')

# Retrieve the first 10 rows
df.head(10)
```

```python
# Lords
df.loc[(df.index >= 160920733001) & (df.index <= 160920733060), "Section"] = "lords-CSE A"
df.loc[(df.index >= 160920733061) & (df.index <= 160920733120), "Section"] = "lords-CSE B"
df.loc[(df.index >= 160920733121) & (df.index <= 160920733179), "Section"] = "lords-CSE C"
df.loc[(df.index >= 160920733301) & (df.index <= 160920733306), "Section"] = "lords-CSE A"
df.loc[(df.index >= 160920733307) & (df.index <= 160920733312), "Section"] = "lords-CSE B"
df.loc[(df.index >= 160920733313) & (df.index <= 160920733319), "Section"] = "lords-CSE C"

# MJ
df.loc[(df.index >= 160420733001) & (df.index <= 160420733060), "Section"] = "mj-CSE A"
df.loc[(df.index >= 160420733061) & (df.index <= 160420733121), "Section"] = "mj-CSE B"
df.loc[(df.index >= 160420733301) & (df.index <= 160420733306), "Section"] = "mj-CSE A"
df.loc[(df.index >= 160420733307) & (df.index <= 160420733312), "Section"] = "mj-CSE B"

# Deccan
df.loc[(df.index >= 160320733001) & (df.index <= 160320733045), "Section"] = "deccan-CSE A"
df.loc[(df.index >= 160320733046) & (df.index <= 160320733090), "Section"] = "deccan-CSE B"
df.loc[(df.index >= 160320733301) & (df.index <= 160320733305), "Section"] = "deccan-CSE A"
df.loc[(df.index >= 160320733306) & (df.index <= 160320733309), "Section"] = "deccan-CSE B"

# Fill NaN values with an appropriate default section or label
df.fillna("Unknown", inplace=True)

# Display a sample of the DataFrame
df.sample()
```

```python
# replacing 'Promoted' with '-1'
df = df.replace('PROMOTED--', "-1")

# Removing the substring 'PASSED-' from all the students that have cleared
df["Overall GPA"] = df["Overall GPA"].str.replace("PASSED-", "")
```

```python
# DataFrame containing students that have cleared all subjects in current sem
current_clear_students_df = compressed_results_df[compressed_results_df["Overall GPA"] != -1]
current_clear_students_df
```

```python
# Assigning Rank based on GPA
current_clear_students_df["Rank"] = current_clear_students_df["Overall GPA"].rank(ascending=False)
current_clear_students_df.head(10)
```

```python
# Students with the highest Rank
current_clear_students_df.sort_values(by="Rank")[:30]
```

```python
# DataFrame Containing students that got backlog
back_students_df = compressed_results_df.loc[compressed_results_df["Overall GPA"] == -1]
```

```python
# Counting number of students that have cleared and have not cleared all subjects 
cleared_count = len(current_clear_students_df)
back_count = len(back_students_df)
print("Total students that cleared all subjects", cleared_count)
print("Total students that did not clear all subjects", back_count)
```

```python
#Passed vs Failed combining lords,mj and deccan
plt.figure(figsize=(7,7))
plt.style.use("_mpl-gallery")
pallet = sns.color_palette("bright")
data = [cleared_count, back_count]
plt.title("Passed vs Failed combining lords,mj and deccan")
plt.pie(data, colors=pallet, autopct='%.0f%%', 
        explode=[0.1, 0.0], pctdistance=0.5 )
plt.legend(["Passed", "Failed"])


plt.show()
```

```python
# Highest GPA per section
max_gpa_per_sec = current_clear_students_df.groupby(by="Section").agg({"Overall GPA": "max"})
current_clear_students_df.merge(max_gpa_per_sec, on=["Overall GPA", "Section"], how="inner").iloc[:,:-1]
```

```python
# Number of students cleared per section
clear_per_section = current_clear_students_df.groupby(by="Section").agg({"Overall GPA": "count"}).rename(columns={"Overall GPA":"Passed"})
clear_per_section
```

```python
# Total students who failed per Section
back_per_section = back_students_df.groupby(by="Section").agg({"Overall GPA": "count"}).rename(columns={"Overall GPA":"Failed"})
back_per_section
```

```python
# Combining the two data frames 
section_data_merged = clear_per_section.join(back_per_section, on="Section")
section_data_merged
```

```python
# Passed vs Failed per section
plt.style.use("_mpl-gallery")
x = np.arange(len(section_data_merged))

plt.figure(figsize=(8, 8))
plt.bar(x-0.2, section_data_merged["Passed"], 0.4, color="#2B2730")
plt.bar(x+0.2, section_data_merged["Failed"], 0.4, color='#73BBC9')
plt.xticks(x, section_data_merged.index)
plt.xlabel("Section")
plt.ylabel("Number of Students")
plt.legend(section_data_merged.columns, title="Passed or failed")
plt.title("Passed vs Failed per section in three colleges")
plt.show()
```
